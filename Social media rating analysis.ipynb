### Encoding quantitave variables into categorical variables

```python
llist=list()
for value in df['Youtube']:
    if value < 3 :
        x='Very Negative'
        llist.append(x)
    elif value >= 3 and value <5:
        x='Negative'
        llist.append(x)
    elif value >=5 and value <7:
        x='Neutral'
        llist.append(x)
    else:
        x='Positive'
        llist.append(x)
df['Youtube_rate']=llist
```

```python
llist2=list()
for value in df['Facebook']:
    if value < 3 :
        x='Very Negative'
        llist2.append(x)
    elif value >= 3 and value <5:
        x='Negative'
        llist2.append(x)
    elif value >=5 and value <7:
        x='Neutral'
        llist2.append(x)
    else:
        x='Positive'
        llist2.append(x)
df['Facebook_rate']=llist2
```

```python
llist3=list()
for value in df['Twitter']:
    if value < 3 :
        x='Very Negative'
        llist3.append(x)
    elif value >= 3 and value <5:
        x='Negative'
        llist3.append(x)
    elif value >=5 and value <7:
        x='Neutral'
        llist3.append(x)
    else:
        x='Positive'
        llist3.append(x)
df['Twitter_rate']=llist3
```

```python
llist4=list()
for value in df['Instagram']:
    if value < 3 :
        x='Very Negative'
        llist4.append(x)
    elif value >= 3 and value <5:
        x='Negative'
        llist4.append(x)
    elif value >=5 and value <7:
        x='Neutral'
        llist4.append(x)
    else:
        x='Positive'
        llist4.append(x)
df['Instagram_rate']=llist4
```

### Exploratory Data analysis

> Note that this data was collected by a survey that asked people to rate these social media platforms:

> Youtube

> Facebook

> Twitter

> Instagram

> The rate is from 0 to 10 where 0 means that this plaform has a negative impact on their mental health and 10 means that it has a postive impact on their mental health

### What is the impact of Youtube on people's mental health ?

```python
plt.style.use('seaborn')
df['Youtube'].hist(bins=10,color='#5F9EA0')
plt.title('Youtube Rating Histogram',fontname='Times new roman',fontsize=14)
plt.xlabel('Rating')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
print('Rate: ,\n',df['Youtube_rate'].value_counts(normalize=True))
df['Youtube_rate'].value_counts(normalize=True).sort_values().plot(kind='barh')
plt.title('Youtube Rating Proportion',fontname='Times new roman',fontsize=14)
```

```python
print('column name: ,\n',df['Youtube_rate'].value_counts())
sns.countplot(x='Youtube_rate', data=df,palette="cubehelix")
plt.title('Youtube Rating',fontname='Times new roman',fontsize=14)
```

#### INSIGHT 1: Youtube seems to have a positive impact on  menatal health as 67% of the partcipant chose that it has positive impact, 22% percent chose that it is neutral, and about 11% of participant fount it negative or very negative.
- Youtube is likely to be a platform that has a good impact on it's users, maybe because of the free educational material that   exist on it, but there is also a negative impact from this platform as it was showed in this survey maybe because of the bad   content that also exist on Youtube
